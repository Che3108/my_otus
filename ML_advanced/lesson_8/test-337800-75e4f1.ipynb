```python
def get_data(f):
    art = f.split('_')[0]
    data = {'metadata' : {'art': art}}
    
    url = 'http://127.0.0.1:8000/send-photo'

    headers = {
        'access_token': '1234567890',
    }

    s = requests.Session()

    s.headers.update(headers)

    with open(f'./main_pics/{f}','rb') as f:
        files = {
            'files': f,
            'json': (None, json.dumps(data), 'application/json'),
        }

        r = s.post(url,files=files, json=data)
        
    return r
    
```

```python
%%time
f = files[0]
r = get_data(f)
r.status_code
```

```python
r.json()
```

```python
%%time

with ThreadPoolExecutor(200) as pool:
    futures = []
    
    for f in files[:200]:
        futures.append(pool.submit(get_data, f=f))
    
    res = []
    
    for f in futures:
        res.append(f.result())
```

```python
def get_generation(text):
    url = 'http://127.0.0.1:8000/complitions'

    headers = {
        'access_token': '1234567890',
    }

    s = requests.Session()

    s.headers.update(headers)

    data = {'text': '    ,    - :'}

    r = s.post(url, json=data)
    
    if r.status_code == 200:
        return r.json()
    else:
        return None
```

```python
texts = ['    ,    - :'] * 6
texts[:4], len(texts)
```

```python
%%time

with ThreadPoolExecutor(3) as pool:
    futures = []
    
    for t in texts:
        futures.append(pool.submit(get_generation, text=t))
    
    res = []
    
    for f in futures:
        res.append(f.result())
```
