```python
import pyarrow.feather as feather
import pandas as pd

import pickle
```

```python
#  ! wget https://raw.githubusercontent.com/a-milenkin/Datasetes_for_Piplines/main/responses.csv
```

```python
df = pd.read_csv('responses.csv')
```

```python
df = pd.concat([df] * 100)
```

```python
df
```

```python
%%time

with open('test_data_pd.pkl', 'wb') as f:
    pickle.dump(df, f)
```

```python
df.to_numpy()
```

```python
%%time

with open('test_data_np.pkl', 'wb') as f:
    pickle.dump({'columns': df.columns,
                'data': df.to_numpy()}, 
                f)
```

```python
%%time
feather.write_feather(df, 'test_data.ftr')
```

```python
%%time
feather.write_feather(df, 'test_data_compr.ftr', compression='zstd')
```

```python
with open('test_data_compr.ftr', 'rb') as f:
    df1 = feather.read_feather(f)
```
