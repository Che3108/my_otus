```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
iris = load_iris()
```

```python
iris
```

```python
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=1)
```

```python
clr1 = RandomForestClassifier()

clr1.fit(X_train, y_train)
```

```python
clr2 = DecisionTreeClassifier(max_depth=5)
```

```python
# help(DecisionTreeClassifier)
```

```python
clr2.fit(X_train, y_train)
```

```python
clr2
```

```python
plot_tree(clr2)
```

```python
# Convert into ONNX format
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType, DoubleTensorType
import skl2onnx
```

```python
dir(skl2onnx)
```

```python
initial_type = [('float_input', DoubleTensorType([None, 4]))]

onx1 = convert_sklearn(clr1, initial_types=initial_type)

onx2 = convert_sklearn(clr2, initial_types=initial_type)
```

```python
# help(convert_sklearn)
```

```python
onx2
```

```python
onx2
```

```python
with open("rf_iris.onnx", "wb") as f:
    f.write(onx1.SerializeToString())
```

```python
from onnx import ModelProto  # noqa: E402


```

```python
content = onx2.SerializeToString()
```

```python
content
```

```python
model = ModelProto()
model.ParseFromString(content)
```

```python
! python3 -m pip install pydot
```

```python
# We convert it into a graph.

from onnx.tools.net_drawer import GetOpNodeProducer, GetPydotGraph  # noqa: E402

pydot_graph = GetPydotGraph(
    model.graph, name=model.graph.name, rankdir="LR", node_producer=GetOpNodeProducer("docstring")
)
```

```python
pydot_graph.write_dot("graph.dot")
```

```python
import os

os.system("dot -O -Tpng graph.dot")
```
