```python
formula = 'y ~ X'

family_GLMG = sm.families.Gaussian()
family_GLMB = sm.families.Binomial()


model_GLMG = glm(formula = formula, data = data, family = family_GLMG).fit()
print(model_GLMG.summary())

model_GLMB = glm(formula = formula, data = data, family = family_GLMB).fit()
print(model_GLMB.summary())
```

```python
# Давате построим модель, на основе GLM, устанавливающую зависимость между площадью комнаты и площадью кухни
```

```python
apart.info()
```

```python
apart = apart[['Площадь кухни, м2', 'Площадь комнаты, м2']]
apart.columns = ['X', 'y']

model_formula = 'y ~ X'

link_function = sm.families.links.identity()
model_family = sm.families.Poisson(link = link_function)

model_fit = glm(formula = model_formula,
                 data = apart,
                 family = model_family).fit()

print(model_fit.summary())
intercept, slope = model_fit.params

print('Intercept =', intercept)
print('Slope =', slope)


print(model_fit.conf_int())
```

```python
apart = pd.read_excel('/content/github.com/ankislyakov/data/raw/main/оценка+стоимости+квартиры.xlsx')
```

```python
apart.columns = ['y', 'X1', 'X2', 'X3', 'X4', 'X5']
```

```python
model_formula = 'y ~ X1 + X2 + X3 + X4'

link_function = sm.families.links.identity()
model_family = sm.families.Gaussian(link = link_function)

model_fit = glm(formula = model_formula,
                 data = apart,
                 family = model_family).fit()

print(model_fit.summary())

print('Params =', model_fit.params)


print(model_fit.conf_int())
```

```python
test = apart.sample(10).reset_index() [['y', 'X1', 'X2', 'X3', 'X4']]

pred_mod= model_fit.predict(test[['X1', 'X2', 'X3', 'X4']])

print(pd.concat([test, pred_mod], axis = 1))
```

Можно потренироваться на сгенерированных данных:

```python
RANDOM_SEED = 42
rng = np.random.default_rng(RANDOM_SEED)

size = 100
true_intercept = 1
true_slope = 2

x = np.linspace(0, 1, size)
# y = a + b*x
true_regression_line = true_intercept + true_slope * x
# add noise
y = true_regression_line + rng.normal(scale=0.5, size=size)

data = pd.DataFrame(dict(x=x, y=y))
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, xlabel="x", ylabel="y", title="GLM model")
ax.plot(x, y, "x", label="исходные данные]")
ax.plot(x, true_regression_line, label="модельные значения", lw=2.0)
plt.legend(loc=0);
```
