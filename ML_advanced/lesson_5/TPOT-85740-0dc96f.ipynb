```python
# импорт библиотек
from tpot import TPOTClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
digits = load_digits()  # загрузка набора изображений рукописных цифр 8х8 пикселей
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target,  train_size=0.75)
```

```python
len(digits.data [0])
```

```python
digits.target
```

```python
tpot = TPOTClassifier(generations=2, verbosity=2, max_time_mins=0.5, random_state=42)
tpot.fit(X_train, y_train)

```

```python
print( tpot.score(X_test, y_test) )
```

```python
tpot.export('tpot_images.py')
```

```python
# еще пример с Ирисами Фишера
from sklearn.datasets import load_iris
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,  train_size=0.75)
```

```python
tpot = TPOTClassifier(generations=2, verbosity=2, max_time_mins=0.5, random_state=42)
tpot.fit(X_train, y_train)
```

```python
print( tpot.score(X_test, y_test) )
tpot.export('tpot_iris.py')
```

```python
# Еще пример - задача регрессии
import pandas as pd

data = pd.read_excel('оценка+стоимости+квартиры.xlsx')
data = data.iloc[:,:5]
data
```

```python
from tpot import TPOTRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,1:5].values, data.iloc[:,0].values,  train_size=0.75)
```

```python
y_train
```

```python
tpot = TPOTRegressor(generations=2, verbosity=2, max_time_mins=0.5, random_state=42)
tpot.fit(X_train, y_train,)
```

```python
print(tpot.score(X_test, y_test))
tpot.export('tpot_kvart.py')
```

```python
preds = tpot.predict(X_test)
```

```python
preds
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, preds)
```
